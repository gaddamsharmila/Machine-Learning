```python
#.DISPLAY IF THE GIVEN NUMBER IS EVEN OR ODD
x=(int(input("Enter the number.")))
if x % 2 ==0:
  print(x,"is even")

else:
  print(x,"is odd")
```

```python
#.PRINT THE SUM OF CURRENT AND PREVIOUS NUMBERS
print("Printing current and previous number and their sum in a range (10)")
previous_num=0
for i in range(1,11):
  sum=previous_num+i
  print("Current number",i,"previous number",previous_num,"Sum:",sum)
  previous_num=i
```

```python
#PRINT THE CHARACTERS FROM THE STRING AT EVEN INDEX
word=input("enter word: ")
print("Original string",word)
size=len(word)
print("Printing only even index characters")
for i in range(0,size-1,2):
  print("index[",i,"]",word[i])
```

```python
def add(*args):
  sum=0
  for i in args:
    sum+=i
    return sum
    print(add(1,2,3))
```

```python
def greet(**kwargs):
  print("hello")
  for key, value in kwargs.items():
    print(value)

greet(name="sharmila",usn="027")
```

```python
import random

gamelist=['rock','paper','scissors']
comp=random.choice(gamelist)
user=input("enter your choice   ")
print(comp)
print(user)

if (comp=="rock" and user=="paper"):
  print("computer won")
elif (comp=="paper" and user=="scissors"):
  print("you won")
elif(comp=="scissors" and user=="paper"):
  print("computer won")

else:
  print("tie")
```
